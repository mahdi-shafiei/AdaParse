```python
import json
from pathlib import Path
```

```python
dir_path = Path('/lus/flare/projects/FoundEpidem/siebenschuh/aurora_gpt/output/nougat/new_nougat/parsed_pdfs')

# loop jsonl
good_pages  = 0
fail_pages  = 0
empty_pages = 0

fail_dict = {}
empty_dict = {}

# 1st: loop jsonl
for jsonl_file in dir_path.glob('*.jsonl'):
    # loop docs
    with open(jsonl_file, 'r') as f:
        for j,line in enumerate(f):
            dd = json.loads(line)

            # doc text
            idx = dd['metadata']['page_char_idx']
            text = dd['text']
            path = dd['path']

            print(path)

            # loop pages
            for page_no in range(len(idx)-1):
                page_text = text[idx[page_no] : idx[page_no+1]]
                if 'MISSING_PAGE_FAIL' in page_text:
                    fail_pages+=1
                    # TODO: append page_no to list fail_dict['path'].append()
                    fail_dict.setdefault(path, []).append(page_no)
                elif 'MISSING_PAGE_EMPTY' in page_text:
                    empty_pages+=1
                    # TODO: append page_no to list empty_dict['path'].append()
                    empty_dict.setdefault(path, []).append(page_no)
                else:
                    good_pages+=1

        #print(path)
        #print(path)
        #if path=='/lus/flare/projects/FoundEpidem/siebenschuh/aurora_gpt/small-pdf-dataset/01_of_20.pdf':
        #    print(jsonl_file)
        #    #break
        #if path=='/lus/flare/projects/FoundEpidem/siebenschuh/aurora_gpt/output/nougat/new_nougat/parsed_pdfs/c8d23375-e7bd-4533-b839-2f686120797e.jsonl':
        #    break
```
